## Modèle de programmation par échange de messages

Chaque processus exécute un sous-programme

- qui est écrit dans un langage classique (C, Fortran, ...),
- qui est généralement le même,
- qui peut avoir des parties dédiées à un processus.

Les variables de chaque sous-programme ont 

- le même nom,
- mais appartiennent à des espaces mémoires différents et correspondent à des données différentes,
- elles sont donc privées.

Un processus $i$ peut communiquer ses données au processus $j$ via des fonctions spéciales d'envoi et de réception.

## Modèles d'exécution

**Modèle d'exécution SPMD**

- Single Program Multiple Data.
- Le même programme est exécuté sur des données différentes.

**Modèle d'exécution MPMD**

- Multiple Program Multiple Data.
- Différents programmes sont exécutés sur des données différentes.

## Concepts d'échange de messages

- Un processus $i$ envoie des données *source* au processus $j$.
- Le processus $j$ reçoit des données du processus $i$ et les met dans *cible*.

<img src="files/figures/echange.png" style="width: 50%;" />

Afin de transmettre correctement un message, certaines informations sont nécessaires

- l'expéditeur,
- le destinataire,
- le type du message,
- sa taille,
- le message.

## MPI

### Présentation générale

- Message Passing Interface: bibliothèque portable, efficace et flexible d'échange de messages,
- conçue en 1993,
- norme définissant une bibliothèque de fonctions, utilisable avec les langages C et Fortran,
- permet d'exploiter des ordinateurs distants ou multiprocesseurs par passage de messages.

### Communications point à point

Communication entre 2 pocessus: un expéditeur et un destinataire.

- La communication a lieu à l'intérieur d'un communicateur.
- Les processus liés à cette communication sont identifiés par leur rang dans ce communicateur. 


Nous commencerons par présenter les communications bloquantes point à point. Une communication est dite bloquante lorsque le processus qui envoie doit attendre que le processus récepteur est complètement reçu le message.

Prenons l'exemple suivant

<img src="files/figures/sendBloc.png" style="width:50%;" />

En Fortran, le code s'écrit 

```
program point_a_point
  implicit none
  include 'mpif.h'
  integer, dimension(MPI_STATUS_SIZE) :: statut
  integer, parameter                  :: etiquette=100
  integer                             :: rang,valeur,code

  call MPI_INIT(code)

  call MPI_COMM_RANK(MPI_COMM_WORLD,rang,code)

  if (rang == 1) then
     valeur=1000
     call MPI_SEND(valeur,1,MPI_INTEGER,5,etiquette,MPI_COMM_WORLD,code)
  elseif (rang == 5) then
     call MPI_RECV(valeur,1,MPI_INTEGER,1,etiquette,MPI_COMM_WORLD,statut,code)
     print *,'Moi, processus 5, j''ai reçu ',valeur,' du processus 1.'
  end if

  call MPI_FINALIZE(code)

end program point_a_point
```

Avec `mpi4py`, le code Python s'écrit

````python
import mpi4py.MPI as mpi

rang = mpi.COMM_WORLD.Get_rank()

if rang == 1:
    valeur = 1000
    mpi.COMM_WORLD.send(valeur, dest = 5)
elif rang == 5: 
    valeur = mpi.COMM_WORLD.recv(source = 1)
    print "Moi, Processus 5, j'ai recu", valeur, "du processus 1."
````

````python
import mpi4py.MPI as mpi
import numpy as np

rank = mpi.COMM_WORLD.Get_rank()
n = 10
if rank == 0:
    sendarray = np.linspace(0., 1., n)
    mpi.COMM_WORLD.Send([sendarray, mpi.DOUBLE], dest = 1)
elif rank == 1:
    recvarray = np.empty(n)
    mpi.COMM_WORLD.Recv([recvarray, mpi.DOUBLE], source = 0)
    print recvarray
````

### Communications point à point non bloquantes

Prenons l'exemple suivant

<img src="files/figures/sendBloc1.png" style="width:50%;" />

````python
import mpi4py.MPI as mpi

rank, size = mpi.COMM_WORLD.Get_rank(), mpi.COMM_WORLD.Get_size()

n = 1000
voisins = [rank - 1, rank + 1]

if voisins[0] == -1: voisins[0] = size - 1
if voisins[1] == size: voisins[1] = 0

recvalue = []
for v in voisins:
    mpi.COMM_WORLD.Send([np.arange(n), mpi.INT], v)
for v in voisins:
    recvalue.append(np.empty(n, dtype=np.int))
    mpi.COMM_WORLD.Recv([recvalue[-1], mpi.INT], v)
print rank, voisins, recvalue
````

Si nous prenons `n` assez petit l'exemple fonctionne. En revanche, pour `n` suffisemment grand (par exemple `n=10000`), nous avons des communications bloquantes. 

MPI fournit différents modes de communication 

- Standards
    
     MPI effectue selon la taille du message une copie temporaire ou non.

    - Si copie: l'envoi se termine lorsque la copie est achevée.
    - Sinon: l'envoi se termine lorsque la réception a commencé.


- Synchrones
 
     L'envoi se termine lorsque la réception a commencé.


- Bufferisées

    Le programmeur effectue une copie temporaire du message. L'envoi se termine lorsque la copie temporaire est achevée.

modes | bloquant | non bloquant 
------|----------|-------------
envoi standard| Send | Isend
envoi synchrone | Ssend | Issend
envoi bufferisé | Bsend | Ibsend
réception | Recv | Irecv

On peut donc corriger l'exemple précédent de la façon suivante

````python
import mpi4py.MPI as mpi

rank, size = mpi.COMM_WORLD.Get_rank(), mpi.COMM_WORLD.Get_size()

n = 1000
voisins = [rank - 1, rank + 1]

if voisins[0] == -1: voisins[0] = size - 1
if voisins[1] == size: voisins[1] = 0

recvalue = []
for v in voisins:
    mpi.COMM_WORLD.Isend([np.arange(n), mpi.INT], v)
for v in voisins:
    recvalue.append(np.empty(n, dtype=np.int))
    mpi.COMM_WORLD.Recv([recvalue[-1], mpi.INT], v)
print rank, voisins, recvalue
````

Il est important d'optimiser les communications : gain en performance.

$\Rightarrow$ Minimiser le temps passé à faire autre chose que des calculs (i.e. l'overhead)

Il existe différentes possibilités

- recouvrir les communications par des calculs,
- éviter la recopie du message dans une zone mémoire temporaire,
- minimiser les surcoûts dus aux appels répétés aux routines de communications.

Si on utilise que des communications non bloquantes, il est nécessaire de pouvoir synchroniser afin de s'assurer que l'on a bien toutes les données nécessaires pour la poursuite du calcul.

La synchronisation peut se faire via la fonction `mpi.Request.Waitall` qui attend que les requêtes passées en argument sous forme de liste Python soient toutes achevées.

````python
import mpi4py.MPI as mpi

rank, size = mpi.COMM_WORLD.Get_rank(), mpi.COMM_WORLD.Get_size()

n = 1000
voisins = [rank - 1, rank + 1]

if voisins[0] == -1: voisins[0] = size - 1
if voisins[1] == size: voisins[1] = 0

recvalue = []
req = []
for v in voisins:
    req.append(mpi.COMM_WORLD.Isend([np.arange(n), mpi.INT], v))
for v in voisins:
    recvalue.append(np.empty(n, dtype=np.int))
    req.append(mpi.COMM_WORLD.Irecv([recvalue[-1], mpi.INT], v))

mpi.Request.Waitall(req)
print rank, voisins, recvalue
````

Reprenons notre classe point en ajoutant une méthode `__add__`.

````python
class point:
    def __init__(self, num, x, y):
        self.num = num
        self.x = x
        self.y = y

    def __str__(self):
        s = "coordonnees du point " + str(self.num) + " :\n"
        s += "x : " + str(self.x) + " , y : " + str(self.y) + "\n"
        return s
    
    def __add__(self, p2):
        return point(0, self.x + p2.x, self.y + p2.y)
````

### Types dérivés

Pour le moment, on sait envoyer et recevoir des données qui ont le même type et qui sont contiguës en mémoire.

La construction de types dérivés dans MPI permet d'envoyer et recevoir une partie d'un tableau, une structure, ... Nous ne traiterons ici que la partie sur les tableaux.

Nous avons une matrice de double et nous voudrions envoyer juste un bloc contigü de celle-ci à un autre processus.

<img src="files/figures/contig.png" style="width:40%;" />

````python
import mpi4py.MPI as mpi
import numpy as np

rank = mpi.COMM_WORLD.Get_rank()

a = rank*np.ones((4, 5))

td = mpi.DOUBLE.Create_contiguous(5)
td.Commit() # validation du nouveau type
mpi.COMM_WORLD.Sendrecv_replace([a[1], td], dest=(rank + 1)%2, source=(rank + 1)%2)
td.Free() # destruction du nouveau type

print a
````

Nous avons une matrice de double et nous voudrions envoyer juste un bloc de celle-ci ayant un pas constant à un autre processus.

<img src="files/figures/pasconst.png" style="width:40%;" />

````python
import mpi4py.MPI as mpi
import numpy as np

rank = mpi.COMM_WORLD.Get_rank()

a = rank*np.ones((4, 5))

td = mpi.DOUBLE.Create_vector(4, 1, 5)
td.Commit()
shape = a.shape
a.shape = (a.size,)
mpi.COMM_WORLD.Sendrecv_replace((a[2:], 1, td), dest=(rank + 1)%2, source=(rank + 1)%2)
td.Free()
a.shape = shape
print a
````

<img src="files/figures/pasconst1.png" style="width:40%;" />
